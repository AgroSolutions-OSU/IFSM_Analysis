```python
## Need to determine at what indexes the table breaks... then we can fully split everything up
# # for this we will be looking specifically for the -----------------------
f = open("IFSM_srm1.txt", 'r')
# get all lines
lines = f.readlines()

# convert the list to a numpy array
test_array = np.array(lines)
 
# find the indices of the value 3 in the array
res_array1 = np.where(test_array == '\t__________________________________________________________________________________\n' )
res_array2 = np.where(test_array =='\t_________________________________________________________________________________\n')
# convert the numpy array back to a list
print(len(res_array1[0]))
print(len(res_array2[0]))

table_sep_idxs = list(res_array1[0]) + list(res_array2[0])
print(table_sep_idxs)
table_sep_idx_sorted = sorted(table_sep_idxs)
# print(res_array[0])
# print(test_array[res_array[0]])

## now take this list and create groups of three...
### groups of three because that is how the tables are formatted essentially with these three lines
table_idxs = []
counter = 0
# get value and the couter
curr_list = 0
for count, val in enumerate(table_sep_idx_sorted , start = counter):
    
    if count == 0:
        # create new list
        table_idxs.append([val])
        print("count:", count, "curr_list:",curr_list)
    elif count%3 == 0:
        # change current_list num
        curr_list = len(table_idxs)
        # create new list
        table_idxs.append([val])
        print("count:", count, "curr_list:",curr_list)
    else:
        # add to current list
        table_idxs[curr_list].append(val)
        print("count:", count, "curr_list:",curr_list)
table_idxs
```

```python
## Need to determine at what indexes the table breaks... then we can fully split everything up
# # for this we will be looking specifically for the -----------------------

# LOGICAL NOT WORKS FOR GETTING RID OF STUFF....

with open("IFSM_srm1.txt", 'r') as f:
    lines = f.readlines()

# Convert the list to a NumPy array
test_array = np.array(lines)

# Define the conditions for the truth series
ts1 = test_array == '\t__________________________________________________________________________________\n' 
ts2 = test_array == '\t_________________________________________________________________________________\n'
# Combine conditions using logical OR
combined_condition = np.logical_or(ts1, ts2)

# Use the bitwise NOT operator (~) to exclude lines matching the conditions
filtered_lines = test_array[~combined_condition]

# Print the filtered lines
print(filtered_lines)

# print(res_array[0])
# print(test_array[res_array[0]])
```
